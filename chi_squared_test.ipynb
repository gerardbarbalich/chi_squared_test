```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv("data/bank-additional-full.csv", sep=";")
df.head()
```

```python
# generate binary values using get_dummies
df = df[["job", "y"]]
df_dum = pd.get_dummies(df, columns=["job"], prefix=["job_"] )
df_dum.head()
```

```python
# group by the y variable to form our groups
df_dum = df_dum.groupby(['y']).sum()
df_dum
```

```python
values = df_dum.values
result = chi2_contingency(values)
chisq, p = result[:2]

print ('chisq = {}, p = {}'.format(chisq, p))
```
